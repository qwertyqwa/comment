```python
import json

import pandas as pd
```

```python
input_file = 'date/annotated_data.csv'

# Читаем все строки и фильтруем пустые
with open(input_file, 'r', newline='', encoding='utf-8') as infile:
    lines = infile.readlines()

# Записываем обратно только непустые строки
with open(input_file, 'w', newline='', encoding='utf-8') as outfile:
    for line in lines:
        if line.strip():
            outfile.write(line)
```

```python
data = pd.read_csv('date/annotated_data.csv')

data = data.drop_duplicates(subset=['id'])

columns_to_drop = ['annotation_id', 'annotator', 'created_at', 'lead_time', 'updated_at']
existing_columns = [col for col in columns_to_drop if col in data.columns]
data.drop(columns=existing_columns, inplace=True)


data = data[['id',
            'rating',
            'comment',
            'categories',
            'Вопрос решен',
            'Нравится качество выполнения заявки',
            'Нравится качество работы сотрудников',
            'Нравится скорость отработки заявок',
            'Понравилось выполнение заявки']]

data['Нравится скорость отработки заявок'] = 0
data['Нравится качество выполнения заявки'] = 0
data['Нравится качество работы сотрудников'] = 0
data['Понравилось выполнение заявки'] = 0
data['Вопрос решен'] = 0
data['Другое'] = 0

data.to_csv('date/annotated_data.csv', index=False)
```

```python
data = pd.read_csv('date/annotated_data.csv')


def assignment(id, com, mean):
    data.loc[data['id'] == id, com] = mean

def annonation(categories, id):

    try: categories = json.loads(categories).get('choices', [])
    except: categories = [categories]


    if 'Нравится скорость отработки заявок' in categories:
        assignment(id, 'Нравится скорость отработки заявок', 1)
    if 'Нравится качество выполнения заявки' in categories:
        assignment(id, 'Нравится качество выполнения заявки', 1)
    if 'Нравится качество работы сотрудников' in categories:
        assignment(id, 'Нравится качество работы сотрудников', 1)
    if 'Понравилось выполнение заявки' in categories:
        assignment(id, 'Понравилось выполнение заявки', 1)
    if 'Вопрос решен' in categories:
        assignment(id, 'Вопрос решен', 1)
    if 'Другое' in categories:
        assignment(id, 'Другое', 1)
    if 'ни к одному из перечисленных' in categories:
        assignment(id, 'Другое', 1)
    

for elm in data.id:

    categories = data.loc[data['id'] == elm, 'categories'].values[0]

    annonation(categories, elm)

data.to_csv('date/annotated_data.csv', index=False)
```
